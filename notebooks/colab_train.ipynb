# YOLOv8 Training Pipeline (Cloud)

This notebook is designed to be run in Google Colab to leverage free GPU resources.
It installs Ultralytics, downloads a sample dataset, trains the model, and saves the weights.

```python
# 1. Install Dependencies
%pip install ultralytics
import ultralytics
ultralytics.checks()
```

```python
# 2. Train YOLOv8n (Nano) on COCO8 (Small demo dataset)
# In a real scenario, you would Replace 'coco8.yaml' with your custom dataset YAML.
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data='coco8.yaml', epochs=20, imgsz=640)

print("Training Complete!")
```

```python
# 3. Validate
metrics = model.val()
```
